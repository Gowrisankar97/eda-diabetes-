```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing csv file as datsframe
df=pd.read_csv("diabetes.csv")
```

```python
#size of the dataset
df.shape
```

```python
#columns in dataset
df.columns
```

```python
#assigning target variable
y=df["Outcome"]
```

```python
#removing target variable
x=df.drop(["Outcome"],axis=1)
```

```python
#checking null values in dataset
x.isnull().sum()
#there is no null value in dataset
```

```python
sns.boxplot(y=x["Age"],x=y)
#boxplot for two parameters
#it shows the data distribution of age taken for training
```

```python
sns.boxplot(y=x["BloodPressure"],x=y)
#it shows the data distribution of bloodpressure taken for training
```

```python
#distribution for glucoslevels in histogram plot
g=x["Glucose"]
sns.distplot(g)
```

```python
sns.pairplot(x)
```
